```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
cab_df = pd.read_csv('Cab_Data.csv')
customer_df = pd.read_csv('Customer_ID.csv')
transaction_df = pd.read_csv('Transaction_ID.csv')
city_df = pd.read_csv('City (1).csv')  # Fixed filename

# Show the first few rows of each to make sure they loaded
print("Cab Data:")
display(cab_df.head())

print("Customer Data:")
display(customer_df.head())

print("Transaction Data:")
display(transaction_df.head())

print("City Data:")
display(city_df.head())

# Check for null values
print("Nulls in Cab Data:")
display(cab_df.isnull().sum())

print("\nNulls in Customer Data:")
display(customer_df.isnull().sum())

print("\nNulls in Transaction Data:")
display(transaction_df.isnull().sum())

print("\nNulls in City Data:")
display(city_df.isnull().sum())

# Check for duplicate rows
print("\nDuplicates in each dataset:")
print("Cab:", cab_df.duplicated().sum())
print("Customer:", customer_df.duplicated().sum())
print("Transaction:", transaction_df.duplicated().sum())
print("City:", city_df.duplicated().sum())

# Check data types
print("\nData types for each dataset:")
print("\nCab:")
print(cab_df.dtypes)
print("\nCustomer:")
print(customer_df.dtypes)
print("\nTransaction:")
print(transaction_df.dtypes)
print("\nCity:")
print(city_df.dtypes)
# Convert Excel-style date to real datetime format
cab_df['Date of Travel'] = pd.to_datetime('1899-12-30') + pd.to_timedelta(cab_df['Date of Travel'], unit='D')

# Convert Population and Users to numeric in city_df
city_df['Population'] = city_df['Population'].str.replace(',', '').astype(int)
city_df['Users'] = city_df['Users'].str.replace(',', '').astype(int)

# Step 1: Merge Cab and Transaction on Transaction ID
cab_trans = pd.merge(cab_df, transaction_df, on='Transaction ID')

# Step 2: Merge with Customer on Customer ID
cab_trans_cust = pd.merge(cab_trans, customer_df, on='Customer ID')

# Step 3: Merge with City on City name
full_data = pd.merge(cab_trans_cust, city_df, on='City', how='left')  # left join to keep all cab rides

# Preview the final dataset
print("Final merged dataset:")
display(full_data.head())
print(f"Shape: {full_data.shape}")

# Imports again (just in case)
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,6)

# How many rides each company completed
sns.countplot(data=full_data, x='Company', order=full_data['Company'].value_counts().index)
plt.title("Total Rides by Company")
plt.xlabel("Cab Company")
plt.ylabel("Number of Rides")
plt.show()

# Revenue comparison
sns.barplot(data=full_data, x='Company', y='Price Charged', estimator=sum, ci=None)
plt.title("Total Revenue by Company")
plt.ylabel("Total Revenue")
plt.xlabel("Cab Company")
plt.show()

# Profit comparison
full_data['Profit'] = full_data['Price Charged'] - full_data['Cost of Trip']
sns.barplot(data=full_data, x='Company', y='Profit', estimator=sum, ci=None)
plt.title("Total Profit by Company")
plt.ylabel("Total Profit")
plt.xlabel("Cab Company")
plt.show()


## Hypotheses & EDA Results

### Hypothesis 1:
**Yellow Cab generates more total profit than Pink Cab.**

‚û°Ô∏è *Result:* True ‚Äî based on the profit chart above, Yellow Cab has significantly higher total profit across the dataset.

---

### Hypothesis 2:
**Pink Cab has more total rides than Yellow Cab.**

‚û°Ô∏è *Result:* False ‚Äî the count plot shows Yellow Cab completed more rides.

---

### Hypothesis 3:
**Average revenue per ride is higher for Pink Cab.**

‚û°Ô∏è## Hypotheses & EDA Results

### Hypothesis 1:
**Yellow Cab generates more total profit than Pink Cab.**

‚úÖ *Result:* True ‚Äî based on the profit chart above, Yellow Cab has significantly higher total profit across the dataset.

---

### Hypothesis 2:
**Pink Cab has more total rides than Yellow Cab.**

‚ùå *Result:* False ‚Äî the count plot shows Yellow Cab completed more rides.

---

### Hypothesis 3:
**Average revenue per ride is higher for Pink Cab.**

üß™ *Result:* See test below.


```

```python
# Add a Month column
full_data['Month'] = full_data['Date of Travel'].dt.month_name()

# Plot number of rides by month
sns.countplot(data=full_data, x='Month', order=pd.date_range("2022-01-01", periods=12, freq="M").strftime('%B'))
plt.title("Number of Cab Rides by Month")
plt.xticks(rotation=45)
plt.ylabel("Number of Rides")
plt.xlabel("Month")
plt.show()
```

```python
# Count number of rides per city
city_ride_counts = full_data['City'].value_counts().reset_index()
city_ride_counts.columns = ['City', 'Rides']

# Merge with city_df to get population
city_pop_compare = pd.merge(city_ride_counts, city_df, on='City', how='left')

# Scatter plot: Population vs Rides
sns.scatterplot(data=city_pop_compare, x='Population', y='Rides')
plt.title("Cab Rides vs. City Population")
plt.xlabel("Population")
plt.ylabel("Number of Rides")
plt.show()
```
